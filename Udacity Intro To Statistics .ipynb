```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# Plotting the data points
X = [1300, 1400, 1600, 1500, 2100, 2300]
Y = [88000, 72000, 94000, 86000, 112000, 98000]
plt.plot(X, Y)
print np.interp(2200, X, Y)
```

```python
# Vertical and horizontal bar charts
plt.bar(X, Y, align='center', alpha=0.5, width=10)
plt.show()
plt.barh(X, Y, align='center', alpha=0.5, height=20)
plt.show()
```

```python
# Histogram 
M = np.random.randint(1, 10 ,1000)
plt.hist(M, 9)
```

```python
# Pie chart
N = np.random.randint(1, 10, 6)
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plt.pie(N, labels=N,  explode=(0, 0, 0, 0, 0.15, 0), autopct='%1d%%', colors=colors)
# Align the axis of the pie chart wrt to the surface 
plt.axis('equal')
```
