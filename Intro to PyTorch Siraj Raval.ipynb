# Steps:
#### Import torch.
#### Declare batch_size, input_size, output_size and hidden unit size.
#### Initalize input and output variables. 
#### Initalize weight variables.
#### Define learning rate.
#### Loop over for a given number of iterations.
#### Calculate predicted output and the loss. 
#### Do a backward pass.
#### Calculate the gradients and update the weights.
#### Mark gradients = 0. 
#### End loop

```python
import torch
```

```python
from torch.autograd import Variable
```

```python
dtype = torch.FloatTensor
```

```python
batch_size = 64
input_size = 1000
output_size = 10
hidden_size = 100
```

```python
x = Variable(torch.randn(batch_size, input_size).type(dtype), requires_grad=False)
y = Variable(torch.randn(batch_size, output_size).type(dtype), requires_grad=False)
```

```python
w1 = Variable(torch.randn(input_size, hidden_size).type(dtype), requires_grad=True)
w2 = Variable(torch.randn(hidden_size, output_size).type(dtype), requires_grad=True)
```

```python
learning_rate = 1e-6
```

```python
for iteration in range(500):
    y_output = x.mm(w1).clamp(0).mm(w2)
    loss = (y - y_output).pow(2).sum()
    print(iteration, loss.data[0])
    loss.backward()
    w1.data -= learning_rate * w1.grad.data
    w2.data -= learning_rate * w2.grad.data
    w1.grad.data.zero_()
    w2.grad.data.zero_()
```
